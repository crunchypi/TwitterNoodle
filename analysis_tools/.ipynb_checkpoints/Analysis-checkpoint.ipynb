```python
# // Hacky module import technique
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import pickle
from random import randint
from packages.pipes import prefabs
```

```python
# // File management. 
MAIN_FOLDER_PREFIX = "Data-"
SUB_FOLDER_PREFIX =  "Sub-"
FILE_SUFFIX = ".txt"
IGNORE = ["template-tweet.txt"]
#TOPICS = ["Art", "Food", "Incident", "IT", "Nature", "Sport"]
TOPICS = ["Art", "Food", "IT", "Nature", "Sport"]

# // Parsing.
NAMEFIELD_START = "<NAME>"
NAMEFIELD_END = "</NAME>"
TEXTFIELD_START = "<TEXT>"
TEXTFIELD_END = "</TEXT>"
```

```python
# // Fake tweets.
class sub(): # // Container for user, ment for Tweet
            def __init__(self):
                self.name = None

class Tweet(): # // Faux
    
    def __init__(self):
        self.id_str = None
        self.user = sub()
        self.text = None
        self.coordinates = None
        self.place = None
```

```python
# // Parsing the text of a textfile.
def get_substring(txt:str, start:str, end:str) -> str:
    txt = txt.split()
    index_start = 0
    index_end = 0
    for i in range(len(txt)):
        if start in txt[i]:
            index_start = i + 1
        if end in txt[i]:
            index_end = i
    return " ".join(txt[index_start:index_end])

# // Convert filepath (text file) to 'tweet'.
def convert_to_tweet(path):
    new_obj = Tweet()

    with open(path, "r") as f:
        content = f.read()
        new_obj.user.name = get_substring(
            txt=content,
            start=NAMEFIELD_START,
            end=NAMEFIELD_END
        )
        new_obj.text = get_substring(
            txt=content,
            start=TEXTFIELD_START,
            end=TEXTFIELD_END
        )
    return new_obj
```

```python
# // Automated data fetch

def fetch_data_by_topic(path, topic):

    def crawler(current_path):
        names = os.listdir(current_path)
        collection = []
        for name in names:
            next_path = f"{current_path}/{name}"
            # // Dive into Data collections.
            if MAIN_FOLDER_PREFIX in name:
                res = crawler(current_path=next_path)
                if res: collection.extend(res)
            # // Dive into topics.
            if SUB_FOLDER_PREFIX in name:
                if topic in name:
                    res = crawler(current_path=next_path)
                    if res: collection.extend(res)
            # // Get objects
            if FILE_SUFFIX in name:
                if name not in IGNORE:
                    obj = convert_to_tweet(path=next_path)
                    collection.append(obj)

        return collection

    return crawler(path)

def fetch_data_all(path:str, topics:list):
    collection = []
    for topic in topics:
        collection.extend(fetch_data_by_topic(path=path, topic=topic))
    return collection
```

```python
# // Extra 
def check_duplicates(objects:list):
    for x in objects:
        for y in objects:
            if x is not y:
                if x.user.name == y.user.name:
                    print(f"Duplicate name: {x.user.name}")
                if x.text == y.text:
                    print(f"Duplicate text: {x.text}")
                    
```

```python
# Assign random user ids with no overlap
def assign_uid(objects:list):
    used_uids = []
    def get_id(used_uids):
        while True:
            new = randint(10000, 99999)
            if new not in used_uids:
                used_uids.append(new)
                return new

    for obj in objects:
        obj.id_str = str(get_id(used_uids=used_uids))
```

```python
# // Save and load datasets
def save(data, filename:str = "./Data"):
    pickle_out = open(filename, "wb")
    pickle.dump(data, pickle_out)
    pickle_out.close()

def load(filename:str = ".Data"):
    pickle_in = open(filename, "rb")
    data = pickle.load(pickle_in)
    pickle_in.close()
    return data
```
